```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(x=df['rooms'], y=df['price'],linestyle='-.', marker='o')
plt.show()
```

```python
difdis=df.groupby('district')['price'].mean()
sns.barplot(x=difdis.index, y=difdis.values)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x=df['district'], y=df['price'], hue=df['rooms'])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='district')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='district', kde=True)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='district', kind='ecdf')
plt.xticks(rotation=90)
plt.show()
```
