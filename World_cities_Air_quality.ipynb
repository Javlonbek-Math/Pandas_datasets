```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
```

```python
df.head(5)
```

```python
df.columns=['City','Region','Country','AirQuality','WaterPollution']
```

```python
df.columns
```

```python
df.head(5)
```

```python
df['Region']=df['Region'].map(lambda x: x.replace('"',''))
```

```python
df['Region']=df['Region'].map(lambda x: x.strip())
```

```python
df['Country']=df['Country'].map(lambda x: x.replace('"',''))
```

```python
df['Country']=df['Country'].map(lambda x: x.strip())
```

```python
df.head(5)
```

```python
df1=pd.pivot_table(df, index=['Country','City'], values=['AirQuality','WaterPollution'], aggfunc='mean')
```

```python
df1
```

```python
df2=df1.copy()
```

```python
df2['AirPollution']=100-df1['AirQuality']
```

```python
df2['WaterQuality']=100-df1['WaterPollution']
```

```python
df2
```

```python
df3=pd.DataFrame(df2.values, index=df2.index, columns=[['Air','Air','Water','Water'],['AirQuality','AirPollution','WaterQuality','WaterPollution']])
```

```python
df3
```

```python
df3.columns.names=['Object','Info']
```
